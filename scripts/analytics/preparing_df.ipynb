```python
import pandas as pd 
from datetime import datetime
```

```python
new_category_men_df = pd.read_excel("../../data/new_category_men_df.xlsx")
new_category_women_df = pd.read_excel("../../data/new_category_women_df.xlsx")
old_category_men_df = pd.read_excel("../../data/old_category_men_df.xlsx")
old_category_women_df = pd.read_excel("../../data/old_category_women_df.xlsx")
```

Adding new columns with labels

```python
def add_labels(dataframe, category, gender):

    dataframe["Category type"] = category
    dataframe["Gender"] = gender

    return dataframe
```

```python
new_category_men_df = add_labels(new_category_men_df, "new category", "man")
new_category_women_df = add_labels(new_category_women_df, "new category", "woman")
old_category_men_df = add_labels(old_category_men_df, "old category", "man")
old_category_women_df = add_labels(old_category_women_df, "old category", "woman")
```

```python
main_df = pd.concat([new_category_men_df, new_category_women_df, old_category_men_df, old_category_women_df])
main_df.index = range(len(main_df))
```

Fixing birth date

```python
def fix_birth_date(date_column):
    date_list=[]

    for date in date_column:
        date_int = datetime.strptime(date, "%b %d, %Y")
        date_list.append(date_int)
        
    return date_list
```

```python
main_df["Born"] = fix_birth_date(main_df["Born"])
```

Adding category label

- Old man's categories: 56kg, 62kg, 69kg, 77kg, 85kg, 94kg, 105kg, +105kg
- New man's categories: 55kg, 61kg, 67kg, 73kg, 81kg, 89kg, 96kg, 102kg, 109kg, +109kg

- Old woman's categories: 48kg, 53kg, 58kg, 63kg, 69kg, 75kg, +75kg
- New woman's categories: 45kg, 49kg, 55kg, 59kg, 64kg, 71kg, 76kg, 81kg, 87kg, +87kg

```python
def return_category_str(category_type, gender, weight):

    # Man in old category
    if (category_type == "old category" and gender == "man"): 

        if (weight < 56): return "56kg"
        if (56 <= weight < 62): return "62kg"
        if (62 <= weight < 69): return "69kg"
        if (69 <= weight < 77): return "77kg"
        if (77 <= weight < 85): return "85kg"
        if (85 <= weight < 94): return "94kg"
        if (94 <= weight < 105): return "105kg"
        if (105 <= weight): return "+105kg"

    # Man in new category
    if (category_type == "new category" and gender == "man"): 

        if (weight < 55): return "55kg"
        if (55 <= weight < 61): return "61kg"
        if (61 <= weight < 67): return "67kg"
        if (67 <= weight < 73): return "73kg"
        if (73 <= weight < 81): return "81kg"
        if (81 <= weight < 89): return "89kg"
        if (89 <= weight < 96): return "96kg"
        if (96 <= weight < 102): return "102kg"
        if (102 <= weight < 109): return "109kg"
        if (109 <= weight): return "+109kg"

    # Woman in old category
    if (category_type == "old category" and gender == "woman"): 

        if (weight < 48): return "48kg"
        if (48 <= weight < 53): return "53kg"
        if (53 <= weight < 58): return "58kg"
        if (58 <= weight < 63): return "63kg"
        if (63 <= weight < 69): return "69kg"
        if (69 <= weight < 75): return "75kg"
        if (75 <= weight): return "+75kg"

    # Woman in new category
    if (category_type == "new category" and gender == "woman"): 

        if (weight < 45): return "45kg"
        if (45 <= weight < 49): return "59kg"
        if (59 <= weight < 55): return "55kg"
        if (55 <= weight < 59): return "59kg"
        if (59 <= weight < 64): return "64kg"
        if (64 <= weight < 71): return "71kg"
        if (71 <= weight < 76): return "76kg"
        if (76 <= weight < 81): return "81kg"
        if (81 <= weight < 87): return "87kg"
        if (87 <= weight): return "+87kg"
```

```python
def category_label(dataframe):
    label_list=[]
    
    for index in range(len(dataframe["B.weight"])):

        category_type = dataframe["Category type"][index]
        gender = dataframe["Gender"][index]
        weight = dataframe["B.weight"][index]

        category_str = return_category_str(category_type, gender, weight)

        label_list.append(category_str)

    return label_list
```

```python
main_df["Category"] = category_label(main_df)
```
